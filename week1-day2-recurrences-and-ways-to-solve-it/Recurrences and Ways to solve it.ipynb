```python
def binary_search(arr, target):
    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1  # target not found


# Example usage
arr = [1, 3, 5, 7, 9, 11]
print(binary_search(arr, 7))  # Output: 3
```

# ⭐ Recurrences

A mathematical description of recursive cost.
(OR)

# ⭐ After solving recurrence:
The final Big-O time complexity.

# Example:

T(n) = T(n-1) + 1
→ After solving → O(n)

T(n) = 2T(n-1)
→ After solving → O(2ⁿ)

2️⃣ Recursion tree (draw tree and count work)

3️⃣ Master Theorem (shortcut formula for divide-and-conquer type)

# Master Theorem

If
T(n) = aT(n/b) + f(n)

where a >= 1, b > 1

---

Then, solution is:

T(n) = (n ^ log(base b) a) . [U(n)]

---

U(n) depends on h(n)

---

h(n) = f(n) / (n ^ log(base b) a)

---

Relation between h(n) and U(n) is

1. If h(n) = n^r , where r > 0. Then U(n) = O(n^r)
2. If h(n) = n^r , where r < 0. Then U(n) = O(1)
3. If h(n) = ((log n) ^ i) , where i >= 0. Then U(n) = O(((log n) ^ i+1) / i+1)

T(n) = n^3 . 1
